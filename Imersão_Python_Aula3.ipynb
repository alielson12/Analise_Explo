```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Nova")
df_chatgpt
```

```python
df_principal = df_principal[["Ativo","Data","Último (R$)","Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Último (R$)":"valor_final", "Var. Dia (%)": "var_dia_pct"}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final' ] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on = 'Ativo', right_on ="Código", how = "left")

df_principal
```

```python
df_principal = df_principal.drop(columns = ["Código"])
df_principal
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde_teorica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal = df_principal.rename(columns={"Qtde. Teórica":"Qtde_teorica"}).copy()
df_principal['Qtde_teorica'] = df_principal['Qtde_teorica'].astype(int)
df_principal
```

```python
df_principal['Resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estavel'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da empresa'])
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['Idade (anos)'].apply(lambda x: 'Acima de 100 anos' if x > 100 else ('Entre 50 e 100 anos' if x >= 50 else 'Abaixo de 50 anos'))
df_principal
```

```python
# Analises

# Calcular a maior variação, menor variação e média das variações
maior_variacao = df_principal['variacao_rs'].max()
menor_variacao = df_principal['variacao_rs'].min()
media_variacao = df_principal['variacao_rs'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu'] ['variacao_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['variacao_rs'].mean()


# Imprimindo os resultados
print(f"Maior\tR$ {maior_variacao:,.2f}")
print(f"Menor\tR$ {menor_variacao:,.2f}")
print(f"Média\tR$ {media_variacao:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_principal = df_principal.drop_duplicates()
df_analise_saldo = df_principal.groupby('Resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='variacao_rs', text='variacao_rs', title='Variação Reais por Resultado')
fig.update_traces(hovertemplate='%{y:,.0f}', texttemplate='%{text:,.0f}', textposition='outside')
fig.update_traces(textfont_color='red')
fig.show()
```

```python
fig = px.pie(df_analise_segmento, values='variacao_rs', names='Segmento', title='Variação de Receita por Segmento')
fig.show()
```

```python
df_analise_cat_idade = df_principal.groupby('Cat_idade')['variacao_rs'].sum().reset_index()
df_analise_cat_idade
```

```python
fig = px.bar(df_analise_cat_idade, x='Cat_idade', y='variacao_rs', text='variacao_rs', title='Variação Reais por Categoria de Idade')
fig.show()
```
